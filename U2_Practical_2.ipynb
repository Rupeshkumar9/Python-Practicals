A Python Program that takes an integer and displays its binary, octal, and hexadecimal representations. 

```python
def number_converter():
    # Get input from user
    try:
        number = int(input("Enter a number: "))
        print(f"The Numaber: {number}")
    except ValueError:
        print("Please enter a valid integer.")
        return
    
    # Convert to different bases
    binary = bin(number)[2:]  # Remove '0b' prefix
    octal = oct(number)[2:]   # Remove '0o' prefix  
    hexadecimal = hex(number)[2:].upper()  # Remove '0x' prefix and convert to uppercase
    
    # Display results
    print(f"\nBinary: {binary}")
    print(f"Octal: {octal}")
    print(f"Hexadecimal: {hexadecimal}")
```

```python
# Run the function
number_converter()
```
