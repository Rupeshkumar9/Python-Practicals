Random names select for cr 

```python
import random
student_names = ["Rohan", "Mahesh", "Priya", "Aditya", "Kirti" ]

print(f"The students: {student_names}")

cr = random.choice(student_names)

print(f"Class CR will be: {cr} \n")


```

```python
k=0
for iterate in student_names:
    if iterate==cr:
        student_names[0], student_names[k] = cr, student_names[0]
    k+=1

print(f"List after shuffling: {student_names}")
```
