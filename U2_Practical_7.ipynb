 A Python Program to list of employee records as dictionaries, sort them by salary and display the sorted list.

```python
employe_list = [

    {"name": 'John', "salary": 45000}, 
    {"name": 'Mary', "salary": 52000}, 
    {"name": 'Alex', "salary": 48000}

               ]

print(f"Unsorted list: {employe_list} \n")

# Sort the original list in place
employe_list.sort(key=lambda x: x["salary"])
print(f"Sorted list by Salary: {employe_list}")
```
