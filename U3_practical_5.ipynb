```python
# Write a Python program that reads two numbers from the user and performs division.
# use try-except to handle the following exceptions:

#     ValueError (if the user enters non-numeric input)
#     ZeroDivisionError (if the second number is zero)
#     Display appropriate error messages.


def divide(a,b):
        try:
             return a/b
    
        except ZeroDivisionError as e:
             return e
```

```python
try:
    a = int(input("Enter a Number: "))
    b = int(input("Enter a Number: "))
    print(divide(a,b))

except ValueError as e:
    print(e)




```
