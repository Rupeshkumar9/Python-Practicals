A Python Program to reverse only te vowels in a given String. Keeping the Other Charactres in their Original Positions.

```python
def reverse_vowels(s):
    vowels = 'aeiouAEIOU'
    s = list(s)  # convert string to list for easy swapping
    i, j = 0, len(s) - 1

    while i < j:
        # Move i forward until we find a vowel
        if s[i] not in vowels:
            i += 1
            continue
        # Move j backward until we find a vowel
        if s[j] not in vowels:
            j -= 1
            continue
        # Swap the vowels
        s[i], s[j] = s[j], s[i]
        i += 1
        j -= 1

    return ''.join(s)
```

```python
# Example usage
text = input("Enter a string: ")
print(f"The String: {text}")
result = reverse_vowels(text)
print("String after reversing vowels:", result)
```
