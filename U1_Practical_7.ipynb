A Python Program to find factorial of a number using recursion.

```python
def recurrsion(Number):
    if Number==1:
        return 1
    else:
        return Number*recurrsion(Number-1)

Number=int(input("Enter a Number: "))
print(f"The Number: {Number}")
print(f"Factorial of {Number} is {recurrsion(Number)}")
```
