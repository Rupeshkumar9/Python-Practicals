A program that reads a text file and counts the number of lines, words, and characters.

```python
def count_file():
    line_count=0
    words_count=0
    characters_count=0

    while (True):
        with open("sample.txt", "r") as fileobj:
            if line_count == 0:
                 print("Sample text file (sample.txt):")
                 for interate in fileobj.readlines():
                        print(interate,end="")
                        line_count+=1
                 print(f"\nLines: {line_count}")

            elif words_count == 0:
                for interate in fileobj.readlines():
                     words_count+=len(interate.split())
                print(f"Words: {words_count}")

            elif characters_count==0:
                 for iterate in list(fileobj.read()):
                      if iterate != ' ' and iterate != "\n":
                           characters_count+=1
                 print(f"Characters: {characters_count}") 
                 
            else:
                 break
        
count_file()
```
