A Python Program to to search for an item in a user-provided list and display the position if found, otherwise print “Item not found.”


```python
def item_search(list_elements):
    item = input("Enter the item to be search: " )
    positon=1
    print(f"Searched: {item}")
    if item not in list_elements:
            print("Item not found")
    
    else:
        for interate in list_elements:
            if interate == item:
                print(f"{item} : Item found at position {positon}")
            positon+=1
     

```

```python
def take_input():
    list_elements = []
    while (True):
        #eg. [12, 45, 33, 21, 88]
        element = input("Enter the list element or Type 'exit' to stop: ")

        if element == "exit":
            break
        else:
               list_elements.append(element)
    
    return list_elements

#Run The Program
item_search(take_input())
```
