Program to calculate factorial, power & power of number

```python
import math 

class calculate:
    def square_root(inp):
        return math.sqrt(inp)

    def fact(inp):
        return math.factorial(inp)

    def power(inp):
        return math.pow(inp)
```

```python
num1 = int(input("Enter a Number: "))
print(f"The Number: {num1}")
print("select any operation: ")
print("1. square root of a Number: ")
print("2. Factorial of a Number: ")
print("3. Power of a Number: \n")

num2 = input("Enter Selection: ")
print(f"{num2} selected")

if num2=="1":
    print(f"square root: {calculate.square_root(num1)}")

elif num2=="2":
    print(f"Factorial: {calculate.fact(num1)}")

elif num2=="3":
    print(f"Power: {calculate.power(num1)}")
```
