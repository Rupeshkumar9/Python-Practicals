A Program to generate the Random Password (length = 18)

```python
import random
import string

def generate_password():
        special_char = [ '%' , '#', '$', '*', '!']
        gener_pass = random.choice(string.ascii_letters).upper()
        while(True):                       
                gener_pass+=str(random.randint(0,9))
                gener_pass+=random.choice(string.ascii_letters)
                if len(gener_pass) == 12:
                        return gener_pass
                gener_pass+=random.choice(special_char)
```

```python
if __name__ == "__main__":
       result = generate_password()
       print(f"Your Generated password: {result}")
       print(f"Password Length: {len(result)}")
```
