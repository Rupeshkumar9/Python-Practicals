A Python program to calculate mean, median and mode using fuction

```python
import statistics

def calculate(data):
    print(data)
    mean = statistics.mean(data)
    median = statistics.median(data)
    mode = statistics.mode(data)
    print("Mean:", mean)
    print("Median:", median)         
    print("Mode:", mode)
```

```python
data = []
while(True):
    var1= input("enter the number or Type 'exit' to stop: ")
    
    if var1=="exit":
           break
    else:           
          data.append(int(var1))

calculate(data)
```
