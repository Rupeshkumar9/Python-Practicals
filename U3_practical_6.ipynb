```python
# Create a Python function read_file(filename) that opens and reads a text file.
# Use try-except-finally to handle:
#     FileNotFoundError if the file does not exist
#     PermissionError if the file cannot be opened, Finally FileNotFoundError block should print "File read attempt completed."


def read_file(filename):
    try:
        with open(filename, "r") as file_obj:
                print(file_obj.read())

    except FileNotFoundError as e:
         print(e)
         
    except PermissionError as e:
         print(e)

    finally:
         print("File read attempt completed.")

    










```
