A Python Program to Read a sentence and display how many times each word appears, ignoring case and punctuation

```python
def count_words(Words_list):
    count_dict={ }
    length=len(Words_list)
    for iterate in Words_list:
            count=0
            iterate = iterate.lower()  #lowercase the word
            for num in range(0,length):
                   if (iterate == Words_list[num].lower()):   #check if the word present in the list
                        count+=1      #Count the repeatated words
                    
            if iterate not in count_dict:   #check key of the dictionary to avoid repetation
                    count_dict[iterate]= count  

    return count_dict
```

```python

string = input("Enter the string: ") # eg. input: Rain rain go away, come again another day.
print(f"The String: {string}")
#split() convert the string words into list elements
result= count_words(string.split()) 
print(result)
```
